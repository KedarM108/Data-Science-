# **Part 1**

```python
import pandas as pd

# Load the dataset into a Pandas dataframe
data = pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Customer%20Spending%20Score.csv")
```

```python
data.describe()
```

```python
data
```

```python
grouped_data = data.groupby('Gender')
summary_stats = grouped_data.agg(['mean', 'median', 'min', 'max', 'std'])[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

```python
print(summary_stats)
```

```python
import pandas as pd
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/IRIS.csv')
```

```python
filtered_df = iris[iris['species'].isin(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])]
```

```python
filtered_df.groupby('species').describe()
```

```python
filtered_df.groupby('species').std()
```

```python
filtered_df.groupby('species').mean()
```
